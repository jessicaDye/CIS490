Given two sorted lists, create a new sorted list in O(n) time.

```python
#Trick is to use 2 ptrs, using i and j looking at locations in  a & b
#and increment the point from which you use the smallest element
alist = [1, 2, 3, 4, 10]
blist = [3, 5, 18]
#clist = [1, 2, 3, 4, 5, 10, 18]
#clist = alist
#alist.extend(blist)
#clist = alist.copy()
#clist.extend(blist)

#two pointers, i and j
i = 0
j = 0
clist = []

#loop
while (i < len(alist) )and (j < len(blist)):
  if alist[i] < blist[j]:
    clist.append(alist[i])
    i += 1

  else:
    clist.append(blist[j])
    j+= 1
#cleanup
while i < len(alist):
  clist.append(alist[i])
  i += 1
while j < len(blist):
  clist.append(blist[j])
  j += 1
print(clist)
```
